## Filtering 
***This cell filters the previous dataframes for the most pertinent columns.***

```python
dfList = [df1x,df2x,df4x,df5x,df6x,df7x,df8x,df9x]
dfListD = [df1x,df2x,df6x,df7x,df8x]#,df9x]
dfListR = [df4x,df5x]#df3x

allTweetsdf = pd.DataFrame()
allTweetsR = pd.DataFrame()
allTweetsD = pd.DataFrame()


allTweetsdf = allTweetsdf.append(dfList)
allTweetsdf = allTweetsdf.reset_index(drop=True)

allTweetsD = allTweetsD.append(dfListD)
allTweetsD = allTweetsD.reset_index(drop=True)

allTweetsR = allTweetsR.append(dfListR)
allTweetsR = allTweetsR.reset_index(drop=True)

pd.set_option('display.max_rows', None)

#allTweetsdf.head(10)
allTweetsR.head(10)
```

## Aggregation 
***This cell aggregates the filtered dataframe into multiple samples, and finds a statistic***

```python
#kenbuck4 douglambourn5
colList = ['nlikes','nreplies','nretweets']

x1 = df1x[colList].sample(frac=0.8)
x1a = df1x[colList].sample(frac=0.8)
x1b = df1x[colList].sample(frac=0.8)
df1y = x1.mean() + x1a.mean() + x1b.mean() // 3
df1y.name = df1x['username'][0]

x2 = df2x[colList].sample(frac=0.8)
x2a = df2x[colList].sample(frac=0.8)
x2b = df2x[colList].sample(frac=0.8)
df2y = x2.mean() + x2a.mean() + x2b.mean() // 3
df2y.name = df2x['username'][0]

x6 = df6x[colList].sample(frac=0.8)
x6a = df6x[colList].sample(frac=0.8)
x6b = df6x[colList].sample(frac=0.8)
df6y = x6.mean() + x6a.mean() + x6b.mean() // 3
df6y.name = df6x['username'][0]

x7 = df7x[colList].sample(frac=0.8)
x7a = df7x[colList].sample(frac=0.8)
x7b = df7x[colList].sample(frac=0.8)
df7y = x7.mean() + x7a.mean() + x7b.mean() // 3
df7y.name = df7x['username'][0]

x8 = df8x[colList].sample(frac=0.8)
x8a = df8x[colList].sample(frac=0.8)
x8b = df8x[colList].sample(frac=0.8)
df8y = x8.mean() + x8a.mean() + x8b.mean() // 3
df8y.name = df8x['username'][0]

#x9 = df9x[colList].sample(frac=0.8)
#df9y = x9[colList].mean()
#df9y.name = df9x['username'][0]


sumD = pd.DataFrame()
sumD = sumD.append(df1y.transpose())
sumD = sumD.append(df2y.transpose())
sumD = sumD.append(df6y.transpose())
sumD = sumD.append(df7y.transpose())
sumD = sumD.append(df8y.transpose())
#sumD = sumD.append(df9y.transpose())

sumD.index.name = 'index'
sumD['name'] = sumD.index

sumD
```
